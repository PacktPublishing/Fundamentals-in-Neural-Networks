## Convolutional 1-Dim

This is a tutorial focusing on one single function *Conv1D* function from *tensorflow* library.

```python
import tensorflow as tf
```

```python
# remark:
# sample size = 2
# dimension of each instance to be 5 by 1
input_shape = (2,5,10) # () <= a tuple 
```

```python
x = tf.random.normal(input_shape)
```

```python
x.shape
```

```python
x
```

```python
y = tf.keras.layers.Conv1D(
    filters=3, kernel_size=2,
    activation='relu',
    input_shape=input_shape[1:])
```

```python
y(x).shape
```

```python
y(x)
```
